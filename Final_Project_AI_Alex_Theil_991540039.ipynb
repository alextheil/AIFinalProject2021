```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/parkinsons/telemonitoring/parkinsons_updrs.data',names=["subject#","age","sex","test_time","motor_UPDRS","total_UPDRS","Jitter(%)","Jitter(Abs)","Jitter:RAP","Jitter:PPQ5","Jitter:DDP","Shimmer","Shimmer(dB)","Shimmer:APQ3","Shimmer:APQ5","Shimmer:APQ11","Shimmer:DDA","NHR","HNR","RPDE","DFA","PPE"],header=None) 

dataset = dataset.drop(columns=['motor_UPDRS'])
dataset = dataset.drop([0])
dataset.head()
```

```python
dataset = dataset.drop(columns=['subject#'])
dataset.shape
```

```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.regression import *
exp_reg101 = setup(data = data, target = 'total_UPDRS', numeric_features=["test_time","sex","Jitter(%)","Jitter(Abs)","Jitter:RAP","Jitter:PPQ5","Jitter:DDP","Shimmer","Shimmer(dB)","Shimmer:APQ3","Shimmer:APQ5","Shimmer:APQ11","Shimmer:DDA","NHR","HNR","RPDE","DFA","PPE"], session_id=123) 
```

```python
best = compare_models(exclude = ['ransac'])
```

```python
final_et = finalize_model(tuned_et)
```

```python
print(final_et)
```

```python
predict_model(final_et)
```

R2 increased to 0.83 because the final_et variable is trained on the complete dataset. We need to predict on the unseen dataset.

```python
unseen_predictions = predict_model(final_et, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions.total_UPDRS, unseen_predictions.Label, 'R2')
```
